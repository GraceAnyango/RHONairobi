```python
import re
import string
from nltk.corpus import stopwords
```

```python
import demoji
```

```python
# Defining functions to clean data
def lower_case(tweet):
    tweet = tweet.lower()
    return tweet
def remove_links(tweet):
    tweet = re.sub(r"https\S+"," ",tweet) #removes weblinks
    tweet = re.sub(r"bit.ly/\S+", " ",tweet) #removes weblinks
    return tweet
def remove_user(tweet):
    tweet = re.sub('(rt\s@[a-z]+[a-z0-9-_]+)', '',str(tweet)) #removes @user information
    tweet = re.sub('(@[a-z]+[a-z0-9-_]+)', '',str(tweet))#removes @user information
    return tweet
def remove_hashtags(tweet):
    tweet = re.sub('(#[a-z]+[a-z0-9-_]+)', '', tweet) # removes the hashtags
    return tweet
def remove_emojis(text):
    """This function removes emojis from a given text string"""
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
def deEmojify(tweet):
    tweet = demoji.replace(tweet, '')
    return tweet
def remove_stopwords(tweet):
    stop_words = stopwords.words('english')
    new_stopwords = ["dm", "check", "kindly", "please", "na" , "ni", "get" , "ya", "hello", "still", "need", "cant", "help",
                 "even","hey", "like", "issue", "whats", "hii", "since", "hi", "kwa", "using", "dont", "assist", "im", "know",
                 "people", "someone", "thanks", "thank", "waiting", "today", "ive", "want", "us", "use", "ama", "without",
                 "trying", "guys", "happening", "lol"]
    stop_words.extend(new_stopwords)
    tweet = tweet.join(word for word in tweet.split () if word not in (stop_words))
    return tweet

def basic_clean(tweet):
    """Returns cleaned data, remove punctuation and numbers"""
    tweet = lower_case(tweet)
#     tweet = tokenization(tweet)
    tweet = remove_user(tweet)
    tweet = remove_links(tweet)
    tweet = remove_hashtags(tweet)
    tweet = remove_emojis(tweet)
    tweet = deEmojify(tweet)
#     tweet = remove_stopwords(tweet)
    tweet = re.sub('['+ string.punctuation+ ']+', '',tweet) # Removes punctuation
    tweet = re.sub('([0-9]+)'," ",tweet) # Removes numbers
    return tweet
```

```python
#creating a new column with the clean tweets
clean_data=[]
for i in df["Text"]:
    clean_data.append(basic_clean(i))

df['Clean_Tweets']=clean_data
```

```python
# Create a function that visualizes the value counts of a column
def plot_bar(df, col)-> None:
    """ A function that returns a plot count of columns"""
    plt.figure(figsize=(12,8))
    sns.countplot(data=df, x = col, order=df[col].value_counts().index)
    plt.title(f"{col} count plot", fontsize=25)
    plt.ylabel("Count", fontsize=15)
    plt.xticks(rotation=45)
    plt.xlabel(f"{col}", fontsize=15)
    plt.show()
```

```python
def bi_variate_analysis(df, column1, column2):
        """This function  creates an sns bi-variate countplot for a given column"""
        plt.figure(figsize=(15,10))
        #set seaborn plotting aesthetics
        sns.set(style='white')
        #create grouped bar chart
        sns.countplot(x=column1, hue=column2, data=df,
                    palette = 'rocket' )
        #add overall title
        plt.title(f'{column1} vs {column2}', fontsize=16)
        #add axis titles
        plt.xlabel(column1)
        plt.ylabel('Count')
        #rotate x-axis labels
        plt.xticks(rotation=45)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Clean_Tweets'], df['Score'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create a TfidfVectorizer object
vectorizer = TfidfVectorizer(lowercase=True)

# fit and transform the training data
X_train = vectorizer.fit_transform(X_train)

# transform the testing data using the same vectorizer
X_test = vectorizer.transform(X_test)
```

#### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
clf = DecisionTreeClassifier(criterion='entropy')

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Predict on the test data
y_pred = clf.predict(X_test)

# Evaluate the model's accuracy
print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score
from scipy.stats import randint
```

```python
# Hyperparameter tuning for decision tree
dtc = DecisionTreeClassifier(random_state=42)
param_dtc = {'criterion': ['gini', 'entropy'],
             'max_depth': randint(1, 10),
             'min_samples_split': randint(2, 10),
             'min_samples_leaf': randint(1, 10)}

dtc_cv = RandomizedSearchCV(dtc, param_distributions=param_dtc, cv=5, random_state=42)
dtc_cv.fit(X_train, y_train)
dtc_pred = dtc_cv.predict(X_test)
dtc_acc = accuracy_score(y_test, dtc_pred)
print("Decision Tree Accuracy after tuning: {:.2f}%".format(dtc_acc*100))
print("Best hyperparameters for Decision Tree:", dtc_cv.best_params_)
```

A decision tree accuracy of 63.90% after tuning is a moderate performance. It indicates that the decision tree model was able to correctly predict the target variable for about two-thirds of the observations. However, this also means that there is significant room for improvement, especially if the model is to be used in a real-world setting where higher accuracy is required.
